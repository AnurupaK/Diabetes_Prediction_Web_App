```python
import pandas as pd  #type: ignore
import numpy as np  #type: ignore
from sklearn.preprocessing import StandardScaler #type: ignore
from sklearn.model_selection import train_test_split #type: ignore
from sklearn.linear_model import LogisticRegression # type: ignore
from sklearn.metrics import accuracy_score # type: ignore
import pickle
import warnings
import matplotlib.pyplot as plt # type: ignore
```

```python
diabetes_df = pd.read_csv(rf'F:\Python\VS CODE_py\Diabetes Prediction app\Dataset\Diabetes_data.csv')
df = diabetes_df.copy()
```

```python
df.head()
```

```python
df['Outcome'].value_counts()
```

```python
df.describe()
```

```python
feature_list = list(df.columns)[0:-1]
print(feature_list)
```

```python
for feature in feature_list:
    plt.scatter(df[feature], df['Outcome'], s=10, color='blue')
    plt.xlabel(f'{feature}')
    plt.ylabel('Outcome')
    plt.title(f'{feature} Vs Outcome')
    plt.show()
```

```python
standard = StandardScaler()
```

```python
X = df.drop(columns='Outcome',axis=1)
Y = df['Outcome']
```

```python
X_std = standard.fit_transform(X)
```

```python
X_new = pd.DataFrame(X_std, columns=X.columns)
```

```python
X_new.head()
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X_new,Y,test_size=0.1, stratify=Y, random_state=3) 
```

```python
print(X_train.shape,X_test.shape,X_new.shape)
```

```python
logR = LogisticRegression()
```

```python
logR.fit(X_train,Y_train)
```

```python
##Accuracy checking of model on training data
print(accuracy_score(logR.predict(X_train),Y_train))
```

```python
##Model evaluation on testing data
print(accuracy_score(logR.predict(X_test),Y_test))
```

```python
##Model dumping or saving
pickle.dump(logR,open('LogR_model.pkl','wb'))
```

```python
##Using the model by loading it
model2 = pickle.load(open('LogR_model.pkl','rb'))
```

```python
print(accuracy_score(model2.predict(X_test),Y_test))
```
